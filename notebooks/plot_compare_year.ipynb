```python
from pathlib import Path

import pandas as pd
```

```python
base_years = range(2002, 2021)
intermediate_year = 2021
model_year = 2021

outputs_dir = Path("../outputs")
```

```python
# read all files
load_profiles = {
    base_year: pd.read_csv(
        outputs_dir
        / f"load_base_{base_year}/load_intermediate_{intermediate_year}/load_base{base_year}_intermediate{intermediate_year}_model{model_year}.csv",
        index_col=0,
    )
    for base_year in base_years
}
```

```python
# extra for base !< intermediate
load_profiles.update({2021: pd.read_csv(outputs_dir / "load_base_2021/load_base2021_model2021.csv", index_col=0)})
```

```python
ercot_load_profiles = pd.DataFrame({year: lp.sum(axis=1) for year, lp in load_profiles.items()})
```

```python
ercot_load_profiles.plot()
```
